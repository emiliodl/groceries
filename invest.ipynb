```python
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv("groceries.csv", header = None)
df.replace(np.nan, 0, inplace = True)

df.head(10)
```

```python
def removeAllzeros(l):
    return list(filter(lambda x: x !=0, l))
    
```

```python
list_all_transactions = []

for index, row in df.iterrows():
    listOfTransaction = row.values.tolist()
    listOfTransaction = removeAllzeros(listOfTransaction)

    list_all_transactions.append(listOfTransaction)

print(len(list_all_transaction))

list_all_transactions[0]
  
```

```python
from mlxtend.preprocessing import TransactionEncoder 
te = TransactionEncoder()
te_ary= te.fit(list_all_transactions).transform(list_all_transactions)
df = pd.DataFrame(te_ary, columns=te.columns_)

df
```

```python
from mlxtend.frequent_patterns import apriori, association_rules

frequent_itemsets = apriori(df, min_support = 0.01, use_colnames=True)
frequent_itemsets.sort_values(by=['support'], ascending = False)

frequent_itemsets
```

```python
from mlxtend.frequent_patterns import association_rules

rules = association_rules(frequent_itemsets, metric = "confidence", min_threshold = 0.5)
rules.sort_values(by=['lift'], ascending = False)

rules
```
